```python
#Подключение необходимых модулей
import tensorflow as tf
import numpy as np
import cv2
from matplotlib import pyplot as plt
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.layers import Dense, Activation, Flatten
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.optimizers import SGD, RMSprop, Adadelta, Adam, Adagrad, Adamax, Nadam, Ftrl
```

```python
#Функция для загрузки и предобработки данных
def preprocess():
    mnist = tf.keras.datasets.mnist
    #Реализуется загрузка и чтение данных методом mnist.load_data(), отделяются входные признаки от выходного
    (train_images, train_labels),(test_images, test_labels) = mnist.load_data()
    #Обучающая и тестовая выборки были нормализованы путем деления на число 255.0, чтобы представить все значения пикселей в диапазоне [0;1]
    train_images = train_images / 255.0
    test_images = test_images / 255.0
    #Выходные признаки переводятся в бинарные векторы при помощи метода to_categorical()
    train_labels = to_categorical(train_labels)
    test_labels = to_categorical(test_labels)
    #Функция preprocess() возвращает преобразованные входные и выходные признаки обучающей и тестовой выборок в виде кортежа
    return train_images, train_labels, test_images, test_labels
```

```python
#Вспомогательная функция для построения графиков
def plot_results(history1, history2, metric1, metric2):
    fig, axs = plt.subplots(1,2)
    fig.suptitle("Отображение метрик {metric1} и {metric2}".format(metric1=metric1, metric2=metric2))
    axs[0].plot(history1.history[metric1])
    axs[0].plot(history1.history[metric2])
    axs[1].plot(history2.history[metric1])
    axs[1].plot(history2.history[metric2])
    plt.show()
```

```python
#Вспомогательная функция для вывод данных (оптимизатор, точность и потери)
def print_results(optimizer, accuracy, loss):
    print("")
    print("Оптимизатор - " + optimizer._name + " Скорость обучения - " + str(np.array([optimizer.learning_rate])[0]))
    print('Точность на тестовых данных:', accuracy)
    print('Потери на тестовых данных:', loss)
```

```python
#Функция для загрузки пользовательского изображения не из датасета и предсказания
#Она принимает два параметра path (путь к файлу) и folder (каталог для загрузки сохраненной модели).
def image_pred(path, folder):
    #Считываем изображение в чёрно белом виде
    image = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
    #Изменяем размер на 28x28
    image = cv2.resize(image, (28,28))
    #Выводим изображение
    plt.imshow(image)
    plt.show()
    image = image / 255.0
    #Метод np.argmax необходим для вычисления индекса максимального значения в массиве
    image = np.expand_dims(image, axis=0)
    model = load_model(folder)
    #Предсказание выполняется при помощи метода predict()
    print("Изображение - " + path + " Предсказанное значение " + str(np.argmax(model.predict(image, verbose=0))))
```

```python
#Функция для построения и настройки параметров компиляции
def create_model(optimizer, folder):
    model = Sequential()
    model.add(Flatten())
    #Создаём два слоя: во входном 256 нейронов и в выходном 10
    model.add(Dense(256, activation='relu'))
    model.add(Dense(10, activation='softmax'))
    #В качестве оптимизатора использовалось значение входного параметра optimizer
    model.compile(optimizer=optimizer,loss='categorical_crossentropy', metrics=['accuracy'])
    history = model.fit(preprocess()[0], preprocess()[1], epochs=5, batch_size=128, validation_data=(test_images, test_labels))
    #Модель сохранялась в каталоге с названием folder
    model.save(folder)
    #При помощи метода evaluate() строим модель, обучаем её и оцением
    test_loss, test_acc = model.evaluate(preprocess()[2], preprocess()[3])
    #Данная функция возвращает модель историю обучения и значения точности и потерь на тестовой выборке
    return history, test_loss, test_acc
```

```python
train_images, train_labels, test_images, test_labels = preprocess()

history1, loss1, accuracy1 = create_model(SGD(learning_rate=0.01), 'model1')
print_results(SGD(learning_rate=0.01), accuracy1, loss1)
history2, loss2, accuracy2 = create_model(RMSprop(learning_rate=0.01), 'model2')
print_results(RMSprop(learning_rate=0.01), accuracy2, loss2)
history3, loss3, accuracy3 = create_model(Adadelta(learning_rate=0.01), 'model3')
print_results(Adadelta(learning_rate=0.01), accuracy3, loss3)
history4, loss4, accuracy4 = create_model(Adam(learning_rate=0.01), 'model4')
print_results(Adam(learning_rate=0.01), accuracy4, loss4)
history5, loss5, accuracy5 = create_model(Adagrad(learning_rate=0.01), 'model5')
print_results(Adagrad(learning_rate=0.01), accuracy5, loss5)
history6, loss6, accuracy6 = create_model(Adamax(learning_rate=0.01), 'model6')
print_results(Adamax(learning_rate=0.01), accuracy6, loss6)
history7, loss7, accuracy7 = create_model(Nadam(learning_rate=0.01), 'model7')
print_results(Nadam(learning_rate=0.01), accuracy7, loss7)
history8, loss8, accuracy8 = create_model(Ftrl(learning_rate=0.01), 'model8')
print_results(Ftrl(learning_rate=0.01), accuracy8, loss8)

plot_results(history1, history2, 'accuracy', 'val_accuracy')
plot_results(history1, history2, 'loss', 'val_loss')
plot_results(history3, history4, 'accuracy', 'val_accuracy')
plot_results(history3, history4, 'loss', 'val_loss')
plot_results(history5, history6, 'accuracy', 'val_accuracy')
plot_results(history5, history6, 'loss', 'val_loss')
plot_results(history7, history8, 'accuracy', 'val_accuracy')
plot_results(history7, history8, 'loss', 'val_loss')
```
